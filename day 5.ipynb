```python
import requests
import json

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text

data = json.loads(response)

#print(data)
```

```python
#取出每一筆資料的「留言人數」

for d in data:
    print(d["commentCount"])
```

```python
#取出每一筆資料的「按讚人數」

for d in data:
    print(d["likeCount"])
```

```python
a = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response_a = a.text
pop = json.loads(response_a)

b = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
response_b = b.text
nonpop  = json.loads(response_b)
```

```python
#計算熱門文章的「平均留言人數」與「平均按讚人數」
comment = []
for i in pop:
    comment.append(i["commentCount"])

avg_comment = (sum(comment)/len(comment))
print(avg_comment)

like = []
for i in pop:
    like.append(i["likeCount"])
avg_like = (sum(like)/len(like))
print(avg_like)
```

```python
#計算非熱門文章的「平均留言人數」與「平均按讚人數」

comment_npop = []
for i in nonpop:
    comment_npop.append(i["commentCount"])

avg_comment_npop = (sum(comment_npop)/len(comment_npop))
print(avg_comment_npop)

like_npop = []
for i in nonpop:
    like_npop.append(i["likeCount"])

avg_like_npop = (sum(like_npop)/len(like_npop))
print(avg_like_npop)
```
