```python
a = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response_a = a.text
pop = json.loads(response_a)

b = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
response_b = b.text
nonpop  = json.loads(response_b)
```

```python
#è¨ˆç®—ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
comment = []
for i in pop:
    comment.append(i["commentCount"])

avg_comment = (sum(comment)/len(comment))
print(avg_comment)

like = []
for i in pop:
    like.append(i["likeCount"])
avg_like = (sum(like)/len(like))
print(avg_like)
```
