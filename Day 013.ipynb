```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://www.ptt.cc/bbs/NBA/index.html"

r = requests.get(url)
soup = bs(r.text)
print(soup.prettify())
```

```python
#印出最新文章的「作者」「標題」「時間」

for d in soup.find(class_ ="r-list-container action-bar-margin bbs-screen").find_all(class_ ="r-ent"):
    print(d.find(class_ = 'title').text, d.find(class_ = 'author').text, d.find(class_ = 'date').text)
```

```python
columns = ['author','title','date']
table=[]

for d in soup.find(class_ ="r-list-container action-bar-margin bbs-screen").find_all(class_ ="r-ent"):
    table.append([d.find(class_ = 'author').text, d.find(class_ = 'title').text.replace('\n',""), d.find(class_ = 'date').text])
    
print(table[0:4])

```

```python
#pd.set_option('display.unicode.ambiguous_as_wide', True)
#pd.set_option('display.unicode.east_asian_width', True)
df = pd.DataFrame(table, columns=columns)
df
```
