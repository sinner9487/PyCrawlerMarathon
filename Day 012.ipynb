```python
url = "https://www.ettoday.net/news/news-list.htm"
import requests
from bs4 import BeautifulSoup as bs

r = requests.get(url)
soup = bs(r.text)
```

```python
print(soup.prettify())
```

```python
#取出今天所有的發文
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('a').text)
```

```python
#如果想要依照類別分類，怎麼儲存會比較好？
for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    print(d.find('span').text)
for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    print(d.find('em').text)
for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    print(d.find('a').text)
    
```

```python
import pandas as pd
columns = ['date', 'tag', 'text']
table = []

for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    table.append([d.find('span').text, d.find('em').text, d.find('a').text])
    
df = pd.DataFrame(table, columns=columns)
df.head()
```

```python
#哪一個類別的文章最多
df.sort_values('tag')
```

```python
df.groupby('tag').count()
```
